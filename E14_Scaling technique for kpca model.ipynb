```python
import pandas as pd
```

```python
train_normal = pd.read_csv('./data/normal_training.csv')
```

```python
df_train = train_normal[(train_normal.simulationRun>=1)].iloc[:500,3:]
```

```python
df_train
```

## Normalizer

```python
from sklearn.preprocessing import Normalizer
```

```python
df_norm = Normalizer().fit(df_train)
```

```python
df_norm = df_norm.transform(df_train)
```

```python
df_norm = pd.DataFrame(df_norm)
```

```python
df_norm
```

## MinMax

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
MM_scl = MinMaxScaler()
```

```python
MM_scl.fit(df_train)
```

```python
df_mm = MM_scl.transform(df_train)
```

```python
df_mm = pd.DataFrame(df_train)
```

```python
df_mm
```

## Robust

```python
from sklearn.preprocessing import RobustScaler
```

```python
rb_scl = RobustScaler().fit(df_train)
```

```python
df_rb = rb_scl.transform(df_train)
```

```python
df_rb = pd.DataFrame(df_rb)
```

```python
df_rb
```

## Standardize

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scl = StandardScaler().fit(df_train)
```

```python
df_std = std_scl.transform(df_train)
```

```python
df_std = pd.DataFrame(df_std)
```

```python
df_std
```

## Power Transformer Scaler

```python
from sklearn.preprocessing import PowerTransformer
```

```python
pt_scl = PowerTransformer().fit(df_train)
```

```python
df_pt = pt_scl.transform(df_train)
```

```python
df_pt = pd.DataFrame(df_pt)
```

```python
df_pt
```
