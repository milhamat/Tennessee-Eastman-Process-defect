this experiment are intend for comparing the manual eigen value and eigen vector and the sklearn module in PCA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
```

```python
boston_dataset = load_boston()
```

```python
boston = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
```

```python
boston
```

```python
boston.describe()
```

```python
boston.info()
```

## Base on the Sklearn Model

Standard scaler [here](https://stackoverflow.com/questions/40758562/can-anyone-explain-me-standardscaler)

```python
# standardize the data
from sklearn.preprocessing import StandardScaler

x = StandardScaler().fit_transform(boston)
x = pd.DataFrame(x, columns=boston_dataset.feature_names)
```

```python
## PCA
from sklearn.decomposition import PCA

pca = PCA()
pcat = pca.fit_transform(x)
```

```python
pcat.shape
```

```python
# eigen vectors
pca.components_
```

```python
# eigen values
pca.explained_variance_
```

```python
pca.components_.shape
```

```python
pca.explained_variance_.shape
```

## Combine Methode

```python
# manual normalization
Xc = np.array((boston - boston.mean(axis=0))/boston.std(axis=0))
Xc = pd.DataFrame(Xc, columns=boston_dataset.feature_names)
```

```python
pca2 = PCA()
pcat2 = pca2.fit_transform(Xc)
```

```python
pcat2.shape
```

```python
# eigen vectors combine methode
pca2.components_
```

```python
# eigen values combine methode
pca2.explained_variance_
```

## Manual Methode

```python
# normalize the data
X = np.array((boston - boston.mean(axis=0))/boston.std(axis=0))
```

```python
# X
```

```python
# calculating the the covariance matrix of the data
Cx = np.cov(X, rowvar=False)
```

```python
L, P = np.linalg.eig(Cx)
```

```python
# eigen vector
P
```

```python
# eigen values
L
```

```python
P.shape
```

```python
L.shape
```
