```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_normal = pd.read_csv('./data/normal_training.csv')
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df_train = train_normal[train_normal.simulationRun==1].iloc[:,3:]
```

```python
df_train.info()
```

```python
x = StandardScaler().fit_transform(df_train)
x = pd.DataFrame(x, columns=df_train.columns)
```

```python
x.shape
```

```python
pca = PCA(n_components = 0.9) 
pcat = pca.fit_transform(x)
```

```python
pcat.shape
```

```python
from hotelling.stats import hotelling_t2
```

```python
T2 = hotelling_t2(pcat)
```

```python
stat, f_val, p_val, s = T2
```
