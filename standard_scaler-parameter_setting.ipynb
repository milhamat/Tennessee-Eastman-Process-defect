```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# import seaborn as sns
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
train_normal = pd.read_csv('./data/normal_training.csv')
```

```python
test_faulty = pd.read_csv('./data/faulty_testing.csv')
```

```python
# data preparing
df_train = train_normal[(train_normal.simulationRun>=1)].iloc[:500,3:]
```

```python
std_scl = StandardScaler().fit(df_train)
df_scl = std_scl.transform(df_train)
df_scl = pd.DataFrame(df_scl)
```

```python
df_scl
```

```python
plt.hist(df_scl)
plt.title('Data Training')
plt.show()
```
